```python
from IPython.display import clear_output as co
import random

class GameBoard():
    def __init__(self):
        self.board = [
            '| |','| |','| |','| |','| |',
            '| |','| |','| |','| |','| |',
            '| |','| |','| |','| |','| |',
            '| |','| |','| |','| |','| |',
            '| |','| |','| |','| |','| |',
        ]
        self.game_won = False
        self.game_lost = False
        self.m_coord = 0

class Player(GameBoard):
    def __init__(self):
        super().__init__()
        self.basket_gotten = False
        self.egg_count = 0


class Monster(Player):
    def __init__(self):
        super().__init__()
            
class Egg(Monster):
    def __init__(self):
        super().__init__()

class Door(Egg):
    def __init__(self):
        super().__init__()

class Game(Door):
    def __init__(self):
        super().__init__()
    def new_board(self):
        self.board = [
            '| |','| |','| |','| |','| |',
            '| |','| |','| |','| |','| |',
            '| |','| |','| |','| |','| |',
            '| |','| |','| |','| |','| |',
            '| |','| |','| |','| |','| |',
        ]

    def show_board(self):
        print('*'*50)
        for i in range(len(self.board)):
            if i == 4:
                print(self.board[i],'\n')
            elif i == 9:
                print(self.board[i],'\n')
            elif i == 14:
                print(self.board[i],'\n')
            elif i == 19:
                print(self.board[i],'\n')
            elif i == 24:
                print(self.board[i],'\n')
            elif i == 25:
                print(self.board[i])
            else:
                print(self.board[i],end="")
        print('*'*50)
    
    def player_start(self):
        first_coord = random.randrange(25)
        for i in range(len(self.board)):
            if i == first_coord:
                self.board[i] = '|P|'
    def monster_start(self):
        first_coord = random.randrange(25)
        for i in range(len(self.board)):
            if i == first_coord:
                if self.board[i] == '|P|':
                    print("Oops BAD LUCK! GOTCHA! Better Luck Next Time! *Burps*")
                    self.game_lost = True
                    break
                else:
                    self.board[i] = '|M|'
                    self.m_coord = i
    def basket_start(self):
        empty = True
        while empty:
            first_coord = random.randrange(25)
            for i in range(len(self.board)):
                if i == first_coord:
                    if self.board[i] == '|P|':
                        break
                    elif self.board[i] == '|M|':
                        break
                    else:
                        self.board[i] = '|B|'
                        empty = False
    def playerMove(self):

        u_input = input("Would You like to go up down left or right: ").lower()
        if u_input == 'up':
            try:
                for i in range(len(self.board)):
                    if self.board[i] == '|P|':
                        if self.board[i-5] == '|M|':
                            print("Only Monsta eats you. You do not eat the Monsta. Run away!")
                            break
                        elif self.board[i-5] == '|E|':
                            print("Congratz you got an Egg!")
                            self.egg_count += 1
                            self.board[i-5] = '|P|'
                            self.board[i] = '| |'
                            break
                        elif self.board[i-5] == '|B|':
                            print("Nice Job! Now collect the eggs!")
                            self.basket_gotten = True
                            self.board[i-5] = '|P|'
                            self.board[i] = '| |'
                            break
                        elif self.board[i-5] == '|D|':
                            print("Congrats! You have Won the Game and avoided the Monster")
                            self.game_won = True
                            break
                        else:
                            self.board[i-5] = '|P|'
                            self.board[i] = '| |'
            except IndexError:
                print('Out of range. Maybe you can only go left down or right')
            
        elif u_input == 'down':
            try:
                for i in range(len(self.board)):
                    if self.board[i] == '|P|':
                        if self.board[i+5] == '|M|':
                            print("Only Monsta eats you. You do not eat the Monsta. Run away!")
                            break
                        elif self.board[i+5] == '|E|':
                            print("Congratz you got an Egg!")
                            self.egg_count += 1
                            self.board[i+5] = '|P|'
                            self.board[i] = '| |'
                            break
                        elif self.board[i+5] == '|B|':
                            print("Nice Job! Now collect the eggs!")
                            self.basket_gotten = True
                            self.board[i+5] = '|P|'
                            self.board[i] = '| |'
                            break
                        elif self.board[i+5] == '|D|':
                            print("Congrats! You have Won the Game and avoided the Monster")
                            self.game_won = True
                            break
                        else:
                            self.board[i+5] = '|P|'
                            self.board[i] = '| |'
            except IndexError:
                print('Out of range. Maybe you can only go left up or right')

        elif u_input == 'left':
            try:
                for i in range(len(self.board)):
                    if self.board[i] == '|P|':
                        if self.board[i-1] == '|M|':
                            print("Only Monsta eats you. You do not eat the Monsta. Run away!")
                            break
                        elif self.board[i-1] == '|E|':
                            print("Congratz you got an Egg!")
                            self.egg_count += 1
                            self.board[i-1] = '|P|'
                            self.board[i] = '| |'
                            break
                        elif self.board[i-1] == '|B|':
                            print("Nice Job! Now collect the eggs!")
                            self.basket_gotten = True
                            self.board[i-1] = '|P|'
                            self.board[i] = '| |'
                            break
                        elif self.board[i-1] == '|D|':
                            print("Congrats! You have Won the Game and avoided the Monster")
                            self.game_won = True
                            break
                        else:
                            self.board[i-1] = '|P|'
                            self.board[i] = '| |'
            except IndexError:
                print('Out of range. Maybe you can only go up down or right')

        elif u_input == 'right':
            try:
                for i in range(len(self.board)):
                    if self.board[i] == '|P|':
                        if self.board[i+1] == '|M|':
                            print("Only Monsta eats you. You do not eat the Monsta. Run away!")
                            break
                        elif self.board[i+1] == '|E|':
                            print("Congratz you got an Egg!")
                            self.egg_count += 1
                            self.board[i+1] = '|P|'
                            self.board[i] = '| |'
                            break
                        elif self.board[i+1] == '|B|':
                            print("Nice Job! Now collect the eggs!")
                            self.basket_gotten = True
                            self.board[i+1] = '|P|'
                            self.board[i] = '| |'
                            break
                        elif self.board[i+1] == '|D|':
                            print("Congrats! You have Won the Game and avoided the Monster")
                            self.game_won = True
                            break
                        else:
                            self.board[i+1] = '|P|'
                            self.board[i] = '| |'
                        break
            except IndexError:
                print('Out of range. Maybe you can only go up down or left')
    def monsterMove(self):

        coord = random.randrange(25)
        moved = False
        while not moved:
            if self.board[coord] == '|P|':
                print("Nom Nom Nom. The Monster has eaten you! Game Over!")
                self.board[coord] = '|X|'
                self.board[self.m_coord] = '| |'
                self.m_coord = coord
                self.game_lost = True
                moved = True
            elif self.board[coord] == '|E|':
                break
            elif self.board[coord] == '|D|':
                break
            elif self.board[coord] == '|B|':
                break
            else: 
                self.board[coord] = '|M|'
                self.board[self.m_coord] = '| |'
                self.m_coord = coord
                moved = True
    def egg_spawn(self):
        eggs_placed = 0
        first_coord = random.randrange(25)
        second_coord = random.randrange(25)
        third_coord = random.randrange(25)
        while eggs_placed < 3:
            for i in range(len(self.board)):
                if i == first_coord:
                    if self.board[i] == '|P|':
                        continue
                    elif self.board[i] == '|M|':
                        continue
                    elif self.board[i] == '|E|':
                        continue
                    else:
                        self.board[i] = '|E|'
                        eggs_placed += 1
                if i == second_coord:
                    if self.board[i] == '|P|':
                        continue
                    elif self.board[i] == '|M|':
                        continue
                    elif self.board[i] == '|E|':
                        continue
                    else:
                        self.board[i] = '|E|'
                        eggs_placed += 1
                if i == third_coord:
                    if self.board[i] == '|P|':
                        continue
                    elif self.board[i] == '|M|':
                        continue
                    elif self.board[i] == '|E|':
                        continue
                    else:
                        self.board[i] = '|E|'
                        eggs_placed += 1
    def door_spawn(self):
        empty = True
        while empty:
            first_coord = random.randrange(25)
            for i in range(len(self.board)):
                if i == first_coord:
                    if self.board[i] == '|P|':
                        continue
                    elif self.board[i] == '|M|':
                        continue
                    else:
                        self.board[i] = '|D|'
                        empty = False


def runGame():
    run = False
    while not run:
        inp = input("Would you like to play a dungeon game? Y or N: ").lower()
        co()
        if inp == 'y':
            print("="*50)
            print("Welcome to the Dungeon Game")
            print("="*50)
            game = Game()
            game.player_start()
            game.monster_start()
            game.basket_start()
            game.show_board()
            while not game.game_won and not game.game_lost:
                if game.basket_gotten == True:
                    game.egg_spawn()
                    game.show_board()
                    while game.egg_count <= 3 and not game.game_won and not game.game_lost:
                        print(f'You have {game.egg_count} eggs. Collect them all to spawn the door and GTFO here')
                        if game.egg_count == 3:
                            game.door_spawn()
                            while not game.game_won and not game.game_lost:
                                game.playerMove()
                                game.monsterMove()
                                co()
                                game.show_board()
                        else:
                            game.playerMove()
                            game.monsterMove()
                            co()
                            game.show_board()
                else:
                    game.playerMove()
                    game.monsterMove()
                    co()
                    game.show_board()
        elif inp == 'n':
            print("Okay ;( goodbye!")
            run = True
        else:
            print("Invalid input")
        if game.game_won:
            print("="*50)
            print("CONGRATZ you have collected all the eggs and outran the Monster!")
            print("="*50)
        if game.game_lost:
            print("="*50)
            print("OOPS. Looks like the Monster ate you. GAME OVER!")
            print("="*50)
        
    
runGame()


                    

                    
```
