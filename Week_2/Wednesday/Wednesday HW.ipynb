```python
from IPython.display import clear_output as co
class Cart:
    def __init__(self,cat):
        self.category = cat

class CartItem(Cart):
    def __init__(self,cat,item,price):
        super().__init__(cat)
        self.item = item
        self.price = price

    def add_item(self,category,item,price):
        self.category = category
        self.item = item
        self.price = price
        
    def remove_item(self,cart,item):
        for i in cart:
            if i.get_item() == item:
                cart.remove(i)
                return cart
            else:
                break
    def show_cart(self):
        print(self.category.title(),"\n",self.item.title(),": $",self.price)
    def get_item(self):
        return self.item
        
def show_instructions():
    print("=" * 50)
    print("""Type 'add' to add item to cart.
        Type 'remove' to add item from cart.
        Type 'show' to view all items in your cart.
        Type 'quit' to exit program and show your list.""")
print("=" * 50)
        
shopping_cart = []
total = 0.00
done = False
while not done:
    show_instructions()
    ask = input("What would you like to do?: ").lower()
    co()
    if ask == 'quit':
        print("=" * 50)
        for i in shopping_cart:
            i.show_cart()
        print("=" * 50)
        print("Total: $", total)
        print("Have a good day! Goodbye!")
        done = True
    elif ask == 'add':
        cat = input('What category, ex. Meat,Fruit,Vegetable?: ').lower()
        item = input("What's the item's name?: ").lower()
        price = float(input("What's the item's price?: "))
        total = total + price
        new_cart = CartItem('','',0.0)
        new_cart.add_item(cat,item,price)
        shopping_cart.append(new_cart)
        for i in shopping_cart:
            i.show_cart()
    elif ask == 'remove':
        for i in shopping_cart:
            i.show_cart()
        item = input("Which item would you like to remove?: ").lower()
        for i in shopping_cart:
            if i.item == item:
                total = total - i.price
        for i in range(len(shopping_cart)):
            shopping_cart[i].remove_item(shopping_cart,item)
        
        
    elif ask == 'show':
        for i in shopping_cart:
            i.show_cart()
    else:
        print("Invalid, please try again.")
        continue
        
        
```
